### Modelo para el espacio de soluciones

El espacio de soluciones está formado por todas las expresiones posibles construidas alternando cifras y operadores sin repetición. Cada solución puede representarse como una secuencia ordenada de elementos.

La estructura de datos que mejor se adapta a este problema es el uso de listas (o tuplas) para representar la expresión, junto con cadenas de texto (strings) para su evaluación.

En concreto:

- Se pueden usar listas para almacenar las cifras disponibles y los operadores disponibles.
- Durante la construcción de una solución, se puede generar una lista que combine ambos elementos de forma alternada.
- Finalmente, esta lista se puede convertir en una cadena de texto para evaluarla mediante la función `eval()`.

Por ejemplo, una solución podría representarse como:

[4, '+', 2, '-', 6, '/', 3, '*', 1]

- Las listas permiten gestionar fácilmente elementos sin repetir (eliminando elementos ya usados).

Inicialmente podría pensarse en usar únicamente cadenas de texto, pero esto dificultaría la gestión de los elementos disponibles y la evitación de repeticiones. Por ello, el uso combinado de listas para la construcción y strings para la evaluación resulta la opción más flexible y adecuada.

### Algoritmo por fuerza bruta

Para resolver el problema por fuerza bruta, la idea es generar todas las expresiones posibles que cumplan las restricciones del enunciado y evaluarlas una a una.

La estrategia seguida es la siguiente:

1. Generar todas las permutaciones posibles de 5 cifras distintas tomadas del conjunto {1,2,3,4,5,6,7,8,9}.
2. Generar todas las permutaciones posibles de los 4 operadores {+, -, *, /}, ya que no se pueden repetir.
3. Combinar ambas permutaciones de forma alternada para construir expresiones del tipo:

   d₁ o₁ d₂ o₂ d₃ o₃ d₄ o₄ d₅

4. Convertir cada expresión en una cadena de texto.
5. Evaluar la expresión con `eval()`.
6. Guardar únicamente los resultados que sean enteros.
7. Finalmente, obtener el valor mínimo, el valor máximo y comprobar qué valores enteros aparecen entre ambos.

Este enfoque es de fuerza bruta porque no intenta descartar soluciones intermedias ni aplicar ninguna heurística: simplemente recorre todo el espacio de soluciones permitido por el problema.

```python
import itertools
import math

def es_entero(valor, tol=1e-9):
    """Comprueba si un valor numérico es entero, permitiendo pequeño error numérico."""
    return abs(valor - round(valor)) < tol

def fuerza_bruta():
    cifras = ['1','2','3','4','5','6','7','8','9']
    operadores = ['+','-','*','/']
    
    resultados_enteros = set()
    expresiones_enteras = []
    total_expresiones = 0
    
    # Permutaciones de 5 cifras distintas
    for nums in itertools.permutations(cifras, 5):
        # Permutaciones de los 4 operadores sin repetición
        for ops in itertools.permutations(operadores, 4):
            # Construcción alternada de la expresión
            expresion = nums[0] + ops[0] + nums[1] + ops[1] + nums[2] + ops[2] + nums[3] + ops[3] + nums[4]
            total_expresiones += 1
            
            try:
                valor = eval(expresion)
                
                # Nos quedamos solo con resultados enteros
                if isinstance(valor, (int, float)) and math.isfinite(valor) and es_entero(valor):
                    valor_entero = int(round(valor))
                    resultados_enteros.add(valor_entero)
                    expresiones_enteras.append((expresion, valor_entero))
            
            except ZeroDivisionError:
                # Aunque en este problema no aparece división por cero,
                # se deja por robustez
                pass
    
    return total_expresiones, resultados_enteros, expresiones_enteras

total_expresiones, resultados_enteros, expresiones_enteras = fuerza_bruta()

print("Número total de expresiones evaluadas:", total_expresiones)
print("Número de resultados enteros distintos:", len(resultados_enteros))
print("Valor mínimo alcanzado:", min(resultados_enteros))
print("Valor máximo alcanzado:", max(resultados_enteros))
print("Primeras 10 expresiones con resultado entero:")
for exp, val in expresiones_enteras[:10]:
    print(exp, "=", val)
```

El algoritmo anterior implementa una búsqueda exhaustiva completa. Recorre todas las expresiones posibles que cumplen las restricciones de no repetición y evalúa su resultado. A partir de los resultados enteros obtenidos, ya es posible responder a las preguntas del problema sobre el mínimo, el máximo y la cobertura de los enteros intermedios.

### Complejidad del algoritmo por fuerza bruta

El algoritmo por fuerza bruta genera todas las expresiones posibles que cumplen las restricciones del problema y evalúa cada una de ellas.

Para ello:

- Se generan todas las permutaciones de 5 cifras distintas tomadas de 9 cifras:


P(9,5) = 9 · 8 · 7 · 6 · 5


- Se generan todas las permutaciones de los 4 operadores sin repetición:


4! = 24


Por tanto, el número total de expresiones evaluadas es:


P(9,5) · 4! = 9 · 8 · 7 · 6 · 5 · 24 = 362880


Para cada expresión, las operaciones que se realizan son:
- construir la cadena,
- evaluarla con `eval()`,
- comprobar si el resultado es entero,
- almacenarlo si corresponde.

Como cada expresión tiene tamaño fijo (5 cifras y 4 operadores), el coste de procesar una expresión es constante, es decir, \(O(1)\).

Por tanto, la complejidad temporal total del algoritmo viene dada por el número total de expresiones generadas:


O(P(9,5) · 4!)


En este problema concreto, como el tamaño está fijado, esta cantidad puede verse como constante. Sin embargo, si generalizamos el problema a \(n\) cifras distintas y \(k\) operadores distintos, el crecimiento del espacio de búsqueda sería de tipo factorial, que es uno de los órdenes de complejidad más costosos desde el punto de vista computacional. Los algoritmos de orden factorial crecen muy rápidamente y resultan poco prácticos cuando el tamaño del problema aumenta. 

Por ello, aunque la fuerza bruta es válida para este caso pequeño, no sería una buena estrategia para versiones más grandes del problema.

En cuanto a la complejidad espacial, el algoritmo necesita almacenar:
- las permutaciones que se van generando,
- el conjunto de resultados enteros,
- y, opcionalmente, algunas expresiones encontradas.

Por tanto, el coste en memoria es mucho menor que el temporal, y depende principalmente del número de resultados que decidamos guardar.

```python
import math

# Número total de expresiones evaluadas por fuerza bruta
total_expresiones = math.perm(9, 5) * math.factorial(4)

print("P(9,5) =", math.perm(9, 5))
print("4! =", math.factorial(4))
print("Total de expresiones =", total_expresiones)

print("\nComplejidad temporal:")
print("O(P(9,5) * 4!)")

print("\nEn este caso concreto el tamaño está fijado,")
print("pero si el problema creciera, el comportamiento sería factorial.")
```

### Algoritmo mejorado

Para mejorar el algoritmo por fuerza bruta, se propone un algoritmo de vuelta atrás (backtracking) con poda.

La idea principal es construir la expresión paso a paso, alternando cifras y operadores, en lugar de generar primero todas las expresiones completas. De esta forma, en cada etapa se puede comprobar si la solución parcial sigue siendo prometedora. Si una rama no interesa, se descarta sin necesidad de completarla.

La solución se construye como una secuencia del tipo:

d₁ o₁ d₂ o₂ d₃ o₃ d₄ o₄ d₅

En cada nivel del árbol de búsqueda:
- se elige una cifra no utilizada o un operador no utilizado,
- se actualiza la expresión parcial,
- se comprueba si merece la pena seguir explorando esa rama.

La mejora respecto a la fuerza bruta consiste en introducir podas. En este trabajo se usan las siguientes:

1. No almacenar expresiones completas, sino solo los resultados enteros encontrados.  
   Esto reduce memoria y trabajo adicional.

2. Evaluación progresiva de la expresión parcial.  
   En lugar de esperar al final, se puede calcular el valor acumulado y detener ramas que ya dejan de ser interesantes.

3. Poda por rango razonable.  
   Si durante la construcción el valor parcial crece demasiado en valor absoluto, se descarta esa rama. Esto no garantiza recorrer todo el espacio teórico, pero sí permite encontrar rápidamente muchos resultados enteros válidos con menor coste.

4. Evitar reconstrucciones innecesarias.  
   Solo se añaden elementos disponibles y se retrocede al terminar cada rama. 

Este algoritmo mejora al de fuerza bruta porque no recorre obligatoriamente todas las expresiones completas. El algoritmo de fuerza bruta genera todas las combinaciones y después las evalúa. En cambio, el algoritmo con vuelta atrás construye soluciones por etapas y puede descartar ramas antes de llegar al final. La capacidad de descartar nodos o ramas que no van a formar parte de una solución útil, siempre que la función de descarte sea sencilla. 

Por tanto, aunque en el peor caso sigue siendo un problema combinatorio, en la práctica el algoritmo mejorado reduce el número de expresiones completas evaluadas y hace la búsqueda más eficiente.

```python
import math

def es_entero(valor, tol=1e-9):
    return abs(valor - round(valor)) < tol

def backtracking_mejorado():
    cifras = ['1','2','3','4','5','6','7','8','9']
    operadores = ['+','-','*','/']
    
    resultados_enteros = set()
    expresiones_validas = []
    nodos_visitados = 0
    expresiones_completas = 0
    
    # Umbral de poda: si el valor parcial se dispara, dejamos de explorar
    LIMITE_ABS = 10_000

    def bt(expr_lista, cifras_restantes, ops_restantes, turno_numero):
        nonlocal nodos_visitados, expresiones_completas
        nodos_visitados += 1
        
        # Si la expresión está completa, la evaluamos
        if len(expr_lista) == 9:
            expresiones_completas += 1
            expr = ''.join(expr_lista)
            try:
                valor = eval(expr)
                if isinstance(valor, (int, float)) and math.isfinite(valor) and es_entero(valor):
                    valor_entero = int(round(valor))
                    resultados_enteros.add(valor_entero)
                    expresiones_validas.append((expr, valor_entero))
            except ZeroDivisionError:
                pass
            return
        
        # Si la longitud es impar, toca operador; si es par, toca cifra
        if turno_numero:
            for i, num in enumerate(cifras_restantes):
                expr_lista.append(num)
                
                # Poda simple: si hay ya una subexpresión evaluable, comprobamos su magnitud
                try:
                    expr_parcial = ''.join(expr_lista)
                    valor_parcial = eval(expr_parcial)
                    if isinstance(valor_parcial, (int, float)) and math.isfinite(valor_parcial):
                        if abs(valor_parcial) <= LIMITE_ABS:
                            bt(expr_lista, cifras_restantes[:i] + cifras_restantes[i+1:], ops_restantes, False)
                    else:
                        bt(expr_lista, cifras_restantes[:i] + cifras_restantes[i+1:], ops_restantes, False)
                except SyntaxError:
                    # Expresión parcial todavía incompleta
                    bt(expr_lista, cifras_restantes[:i] + cifras_restantes[i+1:], ops_restantes, False)
                except ZeroDivisionError:
                    pass
                
                expr_lista.pop()
        else:
            for i, op in enumerate(ops_restantes):
                expr_lista.append(op)
                bt(expr_lista, cifras_restantes, ops_restantes[:i] + ops_restantes[i+1:], True)
                expr_lista.pop()

    # Comenzamos eligiendo una cifra
    bt([], cifras, operadores, True)
    
    return resultados_enteros, expresiones_validas, nodos_visitados, expresiones_completas

resultados_bt, expresiones_bt, nodos_visitados, expresiones_completas = backtracking_mejorado()

print("Nodos visitados:", nodos_visitados)
print("Expresiones completas evaluadas:", expresiones_completas)
print("Número de resultados enteros distintos:", len(resultados_bt))
print("Valor mínimo alcanzado:", min(resultados_bt))
print("Valor máximo alcanzado:", max(resultados_bt))
print("\nPrimeras 10 expresiones enteras encontradas:")
for exp, val in expresiones_bt[:10]:
    print(exp, "=", val)
```

El algoritmo mejorado utiliza vuelta atrás para construir las expresiones por etapas y poda para detener ramas poco prometedoras antes de completarlas. De esta forma, se reduce el número de expresiones completas que es necesario evaluar, lo que lo hace más eficiente en la práctica que la búsqueda exhaustiva pura. Así, podemos ver que la mejora de la vuelta atrás depende de poder descartar ramas que no van a aportar soluciones útiles.

### Complejidad del algoritmo mejorado

El algoritmo mejorado está basado en vuelta atrás (backtracking). Esta técnica construye las soluciones por etapas y al avanzar por el árbol de expansión, algunas ramas pueden descartarse antes de llegar al final si se comprueba que no van a proporcionar una solución útil.

O(9 · 4 · 8 · 3 · 7 · 2 · 6 · 1 · 5)

que equivale a:

O(P(9,5) · 4!) = O(362880)


Si generalizamos el problema, el comportamiento sigue siendo de tipo combinatorio/factorial, ambos algoritmos factoriales crecen muy deprisa y resultan poco prácticos cuando el tamaño aumenta. 

La mejora del algoritmo no está en cambiar el peor caso teórico, sino en reducir el número de nodos explorados en la práctica gracias a la poda. Las mejoras en vuelta atrás dependen de la capacidad de descartar nodos o ramas que no van a formar parte de una solución, siempre que la función de descarte sea sencilla computacionalmente. 

Por tanto, se puede concluir lo siguiente:

- Peor caso temporal: igual orden que la fuerza bruta.
- Caso práctico / promedio: mejor que la fuerza bruta, porque no siempre se recorren todas las ramas.
- Complejidad espacial: depende de la profundidad de la recursión y de las estructuras auxiliares utilizadas. Como la profundidad máxima del árbol es 9, el coste asociado a la pila recursiva es pequeño y acotado en este problema concreto.

En resumen, el algoritmo mejorado no cambia radicalmente el orden de complejidad del peor caso, pero sí reduce el trabajo real al evitar explorar ramas poco prometedoras, por lo que resulta más eficiente en la práctica.

### Generación de datos de entrada aleatorios

Aunque el problema está definido con las cifras del 1 al 9 y los operadores básicos, se puede generar un conjunto de datos de entrada aleatorio para realizar pruebas del algoritmo.

En este caso, se propone generar:
- Un subconjunto aleatorio de 5 cifras distintas tomadas del conjunto {1,...,9}.
- Una permutación aleatoria de los 4 operadores {+, -, *, /}.

De esta forma se obtiene una expresión válida que cumple las restricciones del problema (sin repetición y alternancia).

Este tipo de generación permite probar el comportamiento del algoritmo con diferentes combinaciones de entrada y analizar los resultados obtenidos.

```python
import random

def generar_datos_aleatorios():
    cifras = list(range(1, 10))
    operadores = ['+', '-', '*', '/']
    
    # Seleccionamos 5 cifras distintas aleatorias
    cifras_random = random.sample(cifras, 5)
    
    # Permutación aleatoria de operadores (sin repetición)
    operadores_random = random.sample(operadores, 4)
    
    # Construcción de la expresión alternada
    expresion = ""
    for i in range(4):
        expresion += str(cifras_random[i]) + operadores_random[i]
    expresion += str(cifras_random[4])
    
    return cifras_random, operadores_random, expresion

# Generar ejemplo
cifras_r, ops_r, expr_r = generar_datos_aleatorios()

print("Cifras:", cifras_r)
print("Operadores:", ops_r)
print("Expresión generada:", expr_r)
print("Resultado:", eval(expr_r))
```
